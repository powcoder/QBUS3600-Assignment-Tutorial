```python
import numpy as np
import matplotlib.pyplot as plt

# Initialise RNG, so we get same result everytime
np.random.seed(0)

# Number of training points
m = 50

x = np.linspace(0.0, 1.0, m)

# Function coefficients/parameters
beta0 = 4
beta1 = 1.5
beta2 =  3.2
 
# f values from 2nd order polynomial
f = beta0 + beta1 * x + beta2 * np.power(x,2)  
```

```python
# Generate noisy sample from population
sigma2 = 0.1
y = f + np.random.normal(0, np.sqrt(sigma2), m)
```

```python
fig1 = plt.figure()
plt.plot(x, f, label = "f (Ground Truth)")
plt.scatter(x, y, label = "y (Observed Points)", color = "red")
plt.xlabel("Predictor/Feature Value")
plt.ylabel("Target Value")
plt.title("Underlying Function vs Observations")

plt.legend(loc="upper left")
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn1 = KNeighborsRegressor(n_neighbors=10) #
knn1.fit(x.reshape(-1,1),y.reshape(-1,1))
func1=knn1.predict(x.reshape(-1,1))
```

```python
knn4 = KNeighborsRegressor(n_neighbors=5) #
knn4.fit(x.reshape(-1,1),y.reshape(-1,1))
func4=knn4.predict(x.reshape(-1,1))
```

```python
knn2 = KNeighborsRegressor(n_neighbors=1) #
knn2.fit(x.reshape(-1,1),y.reshape(-1,1))
func2=knn2.predict(x.reshape(-1,1))
```

```python
knn3 = KNeighborsRegressor(n_neighbors=25) #
knn3.fit(x.reshape(-1,1),y.reshape(-1,1))
func3=knn3.predict(x.reshape(-1,1))
```

```python
fig3 = plt.figure()
plt.plot(x, f, label = "f (Ground Truth)",color='green')
plt.scatter(x, y, label = "Observed Points", color = "red")
plt.plot(x, func3, label = "KNN-25",color='y') 
plt.legend()
```

```python
fig4 = plt.figure()
plt.plot(x, f, label = "f (Ground Truth)",color='green')
plt.scatter(x, y, label = "Observed Points", color = "red")
plt.plot(x, func1, label = "KNN-10",color='black') 
plt.legend()
```
